**Problem 2**. Analyze the Carbon_West data file using the following code.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
carbon_west = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Carbon_West.dat")
```

Part a. Construct a histogram of CO2 emissions.

```python
carbon_west
```

```python
plt.figure(figsize=(8, 4))
plt.hist(carbon_west["CO2"], bins=10, edgecolor='black')
plt.title(f'Histogram of CO2 Emissions')
plt.xlabel("CO2 Emissions")
plt.ylabel("Frequency")
plt.show()
```

Part b. Find the mean, median, and standard deviation.

```python
mean_CO2 = carbon_west["CO2"].mean()
median_CO2 = carbon_west["CO2"].median()
std_CO2 = carbon_west["CO2"].std()

print(f"Mean CO2 Emissions: {mean_CO2}")
print(f"Median CO2 Emissions: {median_CO2}")
print(f"Standard Deviation of CO2 Emissions: {std_CO2}")
```

```python
minimum_without_dc = np.min(murders_without_dc["murder"])
q1_without_dc = np.percentile(murders_without_dc["murder"], 25)
median_without_dc = np.median(murders_without_dc["murder"])
q3_without_dc = np.percentile(murders_without_dc["murder"], 75)
maximum_without_dc = np.max(murders_without_dc["murder"])

print(f"Minimum: {minimum_without_dc}")
print(f"First Quartile (Q1): {q1_without_dc}")
print(f"Median (Q2): {median_without_dc}")
print(f"Third Quartile (Q3): {q3_without_dc}")
print(f"Maximum: {maximum_without_dc}")
```

**Problem 4**. Read in the Houses data file using the following code.  The variables include price (thousands of dollars), size (square feet), tax bill (dollars), number of bathrooms, number of bedrooms, and whether the house is new (1 = yes, 0 = no) for 100 home sales in Gainesville, Florida. Let
y = selling price and x = size of home.

```python
houses = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Houses.dat")
```

```python
houses
```

Part a. Construct a scatterplot between square feet and price of a home.

```python
plt.figure(figsize=(8,6))
plt.scatter(houses['size'], houses['price'], alpha=0.5)
plt.title('Scatterplot of Square Feet vs. Price')
plt.xlabel('Square Feet')
plt.ylabel('Price')
plt.show()
```

Part b. Find the correlation between the two variables.

```python
correlation_square_feet_price = houses['size'].corr(houses['price'])
correlation_square_feet_price
```

Part c. Find and plot the regression line on top of the scatterplot.

```python
from sklearn.linear_model import LinearRegression
```

```python
x = houses[['size']]
y = houses['price']

model = LinearRegression()
model.fit(x, y)

slope = model.coef_[0]
intercept = model.intercept_

print(f'Slope (Coefficient): {slope}')
print(f'Intercept: {intercept}')

plt.figure(figsize=(6, 4))
plt.scatter(x, y, alpha=0.7, label='Scatterplot')

plt.plot(x, model.predict(x), color='red', label='Regression Line')

plt.title('Scatter Plot with Regression Line')
plt.xlabel('Size')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

Part d. Create a side by side boxplot of price comparing homes that are new and homes that are not new.

```python
plt.figure(figsize=(8, 6))
houses.boxplot(column='price', by='new', vert=False, grid=False)
plt.title('Side-by-Side Box Plot of Price by New vs Not New Homes')
plt.xlabel('Price (Thousands of Dollars)')
plt.ylabel('New (1 = Yes, 0 = No)')
plt.show()
```

Part e. What is the relationship between size and price.  Justify your answer using the plots created. Do the same for the relationship between price and whether or not a house is new, again using the plot to justify your answer.

There is a strong positive linear relationship between home size and price. The scatterplot shows a clear positive relationship - larger homes generally have higher prices - confirmed by the regression line showing a positive slope. Correlation coefficient is 0.834 which supports this.


New homes is likely to have higher prices than not-new homes with more variability in price for new homes. The side-by-side boxplot shows that new homes have higher median prices and wider range of prices compared to older homes.The interquartile range of new homes is much larger which means that new homes tend to be more expensive on average.
